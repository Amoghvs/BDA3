```python
data=pd.read_csv("C:/Users/User/Dropbox/BDA3/tumor.csv")
y=data["y"]
n=data["N"]

def posterior(y,n,log_alpha_over_beta,log_alpha_plus_beta):
    alpha=np.exp(log_alpha_over_beta+log_alpha_plus_beta)/(1+np.exp(log_alpha_over_beta))
    beta=np.exp(log_alpha_plus_beta)/(1+np.exp(log_alpha_over_beta))
    J=len(y)
    dens=0.0
    for j in range(J):
        dens=dens+np.log(1./sp.special.beta(alpha,beta)*sp.special.beta(alpha+y[j],beta+n[j]-y[j]))
    return dens+(log_alpha_over_beta+2*log_alpha_plus_beta)-2*np.log(1+np.exp(log_alpha_over_beta))-5./2*log_alpha_plus_beta
"""
def posterior2(y,n,log_alpha_over_beta,log_alpha_plus_beta):
    alpha=np.exp(log_alpha_over_beta+log_alpha_plus_beta)/(1+np.exp(log_alpha_over_beta))
    beta=np.exp(log_alpha_plus_beta)/(1+np.exp(log_alpha_over_beta))
    J=len(y)
    dens=0.0
    for j in range(J):
        dens=dens+(np.log(sp.special.gamma(alpha+beta))-np.log(sp.special.gamma(alpha))-np.log(sp.special.gamma(beta))+np.log(sp.special.gamma(alpha+y[j]))+np.log(sp.special.gamma(beta+n[j]-y[j]))-np.log(sp.special.gamma(alpha+beta+n[j])))
    return dens+np.log(alpha)+np.log(alpha)-5./2*np.log(alpha+beta)

def posterior_dens(y,n,alpha,beta):
    dens=[]
    for i in range(len(alpha)):
        grid=[]
        for j in range(len(beta)):
            grid.append(posterior2(y,n,alpha[i],beta[j]))
        dens.append(grid)
    return dens
"""
def summ(x):
    percentile=np.percentile(x,[2.5,25,50,75,97.5]).tolist()
    summ=[np.mean(x),np.std(x)]
    summ.extend(percentile)
    summ=DataFrame(summ,index=["mean","sd","2.5%","25%","50%","75%","97.5%"],columns=["info"]).T
    return summ
```
