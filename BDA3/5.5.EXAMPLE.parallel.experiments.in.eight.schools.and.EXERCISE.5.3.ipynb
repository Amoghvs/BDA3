```python
%matplotlib inline
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
from pandas import DataFrame, Series
from numpy import nan as NA
```

```python
data=DataFrame([["A","B","C","D","E","F","G","H"],[28,8,-3,7,-1,1,18,12],[15,10,16,11,9,11,10,18]],index=["School","Estimated treatment effect, y_j","Standard error of effect estimate, sigma_j"]).T
print data
```

```python
def mu_hat(y,sigma,tau):
    J=len(y)
    value1=0.0
    value2=0.0
    for j in range(J):
        value1=value1+1./(sigma[j]**2+tau**2)*y[j]
        value2=value2+1./(sigma[j]**2+tau**2)
    return value1/value2
        
def V_mu(sigma,tau):
    J=len(sigma)
    value=0.0
    for j in range(J):
        value=value+1./(sigma[j]**2+tau**2)
    return 1./value

def p_of_tau(y,sigma,tau):
    J=len(y)
    value=1.0
    for j in range(J):
        value=value*np.power(sigma[j]**2+tau**2,-0.5)*np.exp(-((y[j]-mu_hat(y,sigma,tau))**2)/(2*(sigma[j]**2+tau**2)))
    return np.sqrt(V_mu(sigma,tau))*value

def MH(y,sigma,N):
    mcmc=[]
    para2=np.random.uniform()
    for n in range(N):
        para1=para2+np.random.uniform(-10,50)
        u=np.random.uniform()
        ratio=p_of_tau(y,sigma,para1)/p_of_tau(y,sigma,para2)
        if ratio>1:
            mcmc.append(para1)
        elif ratio>u:
            mcmc.append(para1)
    return np.array(mcmc)[np.array(mcmc)>=0]
            
```

```python
y=data.ix[:,1]
sigma=data.ix[:,2]
tau=MH(y,sigma,10000)
```

```python
plt.plot(tau)
```

```python
plt.hist(tau,bins=50)
```

```python
rang = np.arange(0,30,0.01)
dens=p_of_tau(y,sigma,rang)
plt.plot(rang,dens)
```

```python
mu=np.random.normal(mu_hat(y,sigma,tau),np.sqrt(V_mu(sigma,tau)))
```

```python
def theta_j_hat(y,sigma,mu,tau):
    return (tau**2*y+sigma**2*mu)/(tau**2+sigma**2)

def V_j(sigma,tau):
    J=len(sigma)
    value=[]
    for j in range(J):
        value.append(sigma[j]**2*tau**2/(sigma[j]**2+tau**2))
    return value

def theta_j_vs_tau(y,sigma,mu,tau):
    value1=[]
    for k in range(len(mu)):
        value2=[]
        for i in range(len(tau)):
            value2.append(theta_j_hat(y,sigma,mu[k],tau[i]))
        value1.append(value2)
    return DataFrame(value1)
        
```

```python
theta_j=[]
for i in range(len(data)):
    theta_j.append(theta_j_vs_tau(y[i],sigma[i],mu,rang))
```

```python
E_of_theta_j=[]
for j in range(len(data)):
    E_of_theta_j.append(np.mean(theta_j[j]))
```

```python
plt.xlabel("tau")
plt.ylabel("Posterior Treatment Effects")
plt.ylim([-5,30])
for i in range(len(data)):
    plt.plot(rang,E_of_theta_j[i],label=data.ix[i,0])
plt.legend(bbox_to_anchor=(1.2, 1.0), borderaxespad=0)
```

```python
#analytic ver.ver 

def Expectation_of_theta_j(y,sigma,tau):
    J=len(y)
    value=[]
    for j in range(J):
        value.append((tau**2*y[j]+sigma[j]**2*mu_hat(y,sigma,tau))/(sigma[j]**2+tau**2))
    return value

def std_of_theta_j(y,sigma,tau):
    J=len(y)
    value=[]
    for j in range(J):
        value.append(sigma[j]**2*tau**2/(sigma[j]**2+tau**2)+(sigma[j]**2/(sigma[j]**2+tau**2))**2*V_mu(sigma,tau))
    return np.sqrt(value)
```

```python
E_of_theta_j=Expectation_of_theta_j(y,sigma,rang)
std_of_theta_j=std_of_theta_j(y,sigma,rang)
```

```python
#figure 5.6
plt.xlabel("tau")
plt.ylabel("Posterior Treatment Effects")
plt.ylim([-5,30])
for i in range(len(data)):
    plt.plot(rang,E_of_theta_j[i],label=data.ix[i,0])
plt.legend(bbox_to_anchor=(1.2, 1.0), borderaxespad=0)
```

```python
#figure 5.7
plt.xlabel("tau")
plt.ylabel("Posterior Standard Diviations")
plt.ylim([0,20])
for i in range(len(data)):
    plt.plot(rang,std_of_theta_j[i],label=data.ix[i,0])
plt.legend(bbox_to_anchor=(1.2, 1.0), borderaxespad=0)
```

```python
def sampling_theta_j(y,sigma,mu,tau):
    theta_j_sim=[]
    for j in range(len(data)):
        value=[]
        for i in range(len(tau)):
            value.append(np.random.normal(theta_j_hat(y[j],sigma[j],mu,tau)[i],np.sqrt(V_j(sigma,tau)[j][i])))
        theta_j_sim.append(value)
    return theta_j_sim

theta_j_sim=sampling_theta_j(y,sigma,mu[-200:],tau[-200:])
```

```python
def summ(x,label):
    percentile=np.percentile(x,[2.5,25,50,75,97.5]).tolist()
    summ=[np.mean(x),np.std(x)]
    summ.extend(percentile)
    summ=DataFrame(summ,index=["mean","sd","2.5%","25%","50%","75%","97.5%"],columns=[label]).T
    return summ

#Table 5.3
for j in range(len(data)):
    print summ(theta_j_sim[j],data.ix[j,0])
```
