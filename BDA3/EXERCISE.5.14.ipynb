```python
%matplotlib inline
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
from pandas import DataFrame, Series
from numpy import nan as NA
```

```python
data=DataFrame()

data=data.append(Series([16,9,10,13,19,20,18,17,35,55]),ignore_index=True)
data=data.append(Series([58,90,48,57,103,57,86,112,273,64]),ignore_index=True)
data=data.append(data.ix[0,:]+data.ix[1,:],ignore_index=True)
data=data.T
print data
```

```python
y=data[0]
```

```python
def posterior(y,alpha,beta):
    J=len(y)
    dens=1.0
    for j in range(J):
        dens=dens*np.power(beta,alpha)/np.power(1+beta,y[j]+alpha)*sp.special.gamma(y[j]+alpha)/sp.special.gamma(alpha)
    return dens

def posterior_dens(y,alpha,beta):
    dens=[]
    for i in range(len(alpha)):
        grid=[]
        for j in range(len(beta)):
            grid.append(posterior(y,alpha[i],beta[j]))
        dens.append(grid)
    return dens

def summ1(x,y,label):
    percentile=np.percentile(x,[2.5,25,50,75,97.5]).tolist()
    summ=[y,np.mean(x),np.std(x)]
    summ.extend(percentile)
    summ=DataFrame(summ,index=["raw","mean","sd","2.5%","25%","50%","75%","97.5%"],columns=[label]).T
    return summ

def summ2(x,label):
    percentile=np.percentile(x,[2.5,25,50,75,97.5]).tolist()
    summ=[np.mean(x),np.std(x)]
    summ.extend(percentile)
    summ=DataFrame(summ,index=["mean","sd","2.5%","25%","50%","75%","97.5%"],columns=[label]).T
    return summ
```

```python
alpha_grid=np.linspace(0.001,15,500)
beta_grid=np.linspace(0.001,1,500)
alpha_mesh,beta_mesh = np.meshgrid(alpha_grid,beta_grid)
dens=posterior(y,alpha_mesh,beta_mesh)
```

```python
plt.xlabel("alpha")
plt.ylabel("beta")
plt.contour(alpha_grid,beta_grid,dens)
```

```python
alpha_dens=DataFrame(dens)
beta_dens=DataFrame(dens,index=alpha_grid).T

alpha=np.random.choice(np.array(alpha_grid),p=np.array(np.sum(alpha_dens))/np.sum(np.sum(alpha_dens)),size=400)
beta=[]
for i in range(len(alpha)):
    beta.append(np.random.choice(beta_grid,p=np.array(beta_dens[alpha[i]])/np.sum(beta_dens[alpha[i]])))
```

```python
plt.scatter(alpha,beta)
```

```python
theta=[]
for j in range(len(y)):
    value=[]
    for i in range(len(alpha)):
        value.append(np.random.gamma(y[j]+alpha[i],1)/(1+beta[i]))
    theta.append(value)
```

```python
for j in range(len(theta)):
    Series(theta[j]).plot(kind="kde")
```
