```python
data=pd.read_csv("C:/Users/User/Dropbox/BDA3/football.csv")
```

```python
plt.scatter(data["spread"],data["favorite"]-data["underdog"])
plt.xlim([-1,22])
plt.xlabel("point spread")
plt.ylabel("outcome")
```

```python
d=data["favorite"]-data["underdog"]-data["spread"]

plt.scatter(data["spread"],d)
plt.xlim([-1,22])
plt.xlabel("point spread")
plt.ylabel("outcome - point spread")
```

```python
plt.hist(d,bins=50)
rang=np.arange(-60,60,0.01)
plt.plot(rang,sp.stats.norm.pdf(rang,0,14)*5000,color="r")
```

```python
plt.subplot(121)
plt.scatter(data["spread"],d)
plt.xlim([-1,22])

for i in range(1):
    y_rep=np.random.normal(data["spread"],14)
    pltinfo=122+i
    plt.subplot(pltinfo)
    plt.xlim([-1,22])
    plt.scatter(data["spread"],y_rep-data["spread"])
```

```python
def T(outcome,spread):
    T=[]
    for i in range(51):
        n=1.*i/2
        T.append(np.var(outcome[spread==n]-spread[spread==n]))
    return T
```

```python
T_xy=T(d,data["spread"])
plt.plot(T_xy)
```

```python
T_xy_rep=[]
for i in range(1000):
    T_xy_rep.append(np.var(np.random.normal(1,14,size=1000)-1))
```
