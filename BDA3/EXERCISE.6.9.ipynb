```python
%matplotlib inline
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
from pandas import DataFrame, Series
from numpy import nan as NA
```

```python
data=pd.read_csv("C:/Users/User/Dropbox/BDA3/tumor.csv")
y=data["y"]
n=data["N"]
y_over_n=1.*y/n

def posterior(y,n,log_alpha_over_beta,log_alpha_plus_beta):
    alpha=np.exp(log_alpha_over_beta+log_alpha_plus_beta)/(1+np.exp(log_alpha_over_beta))
    beta=np.exp(log_alpha_plus_beta)/(1+np.exp(log_alpha_over_beta))
    J=len(y)
    dens=0.0
    for j in range(J):
        dens=dens+np.log(1./sp.special.beta(alpha,beta)*sp.special.beta(alpha+y[j],beta+n[j]-y[j]))
    return dens+(log_alpha_over_beta+2*log_alpha_plus_beta)-2*np.log(1+np.exp(log_alpha_over_beta))-5./2*log_alpha_plus_beta

def summ(x):
    percentile=np.percentile(x,[2.5,25,50,75,97.5]).tolist()
    summ=[np.mean(x),np.std(x)]
    summ.extend(percentile)
    summ=DataFrame(summ,index=["mean","sd","2.5%","25%","50%","75%","97.5%"],columns=["info"]).T
    return summ
```

```python
log_alpha_over_beta_grid=np.linspace(-2.5,-1,500)
log_alpha_plus_beta_grid=np.linspace(1.5,3,500)

X,Y =np.meshgrid(log_alpha_over_beta_grid,log_alpha_plus_beta_grid)
dens=posterior(y,n,X,Y)
```

```python
T_rep_max=[]
T_rep_min=[]
T_rep_mean=[]
T_rep_std=[]

for i in range(1000):
    y_over_n_rep=[]
    for j in range(len(y)):
        theta_j=np.random.beta(alpha[i]+y[j],beta[i]+n[j]-y[j])
        y_over_n_rep.append(1.*np.random.binomial(n[j],theta_j)/n[j])
    T_rep_max.append(np.max(y_over_n_rep))
    T_rep_min.append(np.min(y_over_n_rep))
    T_rep_mean.append(np.mean(y_over_n_rep))
    T_rep_std.append(np.std(y_over_n_rep))
```

```python
plt.hist(T_rep_max,bins=15)
plt.plot([np.max(y_over_n),np.max(y_over_n)],[0,300],color="r")
p_value=1.*len(np.array(T_rep_max)[np.array(T_rep_max)>=np.max(y_over_n)])/len(T_rep_max)
print p_value
```
